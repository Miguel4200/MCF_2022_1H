# Método Montecarlo

## Probabilidad y momentos de una variable aleatoria continua

Sea $X$ una variable aleatoria continua, cuya función de densidad está dada por $f(x)$ y definida en un intervalo $[\alpha,\beta]$. Por nuestras clases de probabilidad, sabemos que la probabilidad de que la variable aleatoria $X$, tome un valor menor $x\in[k,n]$ con $k,n\in[\alpha,\beta]$, dentro de su soporte, está dada por:

$$\mathbb{P}(k\leq X\leq n) = \int_{k}^{n} f(x)dx$$

Además, el n-ésimo momento de dicha variable aleatoria $X$, está dado por:

$$\mathbb{E}\left[X^{n}\right] = \int_{\alpha}^{\beta}x^{n}f(x)dx$$

Y, por el teorema del estadístico inconsciente, podemos calcular la esperanza de una función contínua $g$ de dicha variable aleatoria $X$

Sea $g$ una función contínua tal que $g: R\mapsto [\alpha,\beta]$, entonces:

$$\mathbb{E}[g(X)] = \int_{\alpha}^{\beta}g(x)f(x)dx$$

Asimismo se puede definir la esperanza de una función continua de una variable aleatoria:

## Ley fuerte de los grandes números

Sea $\{X_{1}, X_{2}, X_{3},\dots\}$ una sucesión infinita de variables aleatorias independientes e idénticamente distribuidas con media finita (i.e $ \forall i\in\mathbb{N},\quad \mu = \mathbb{E}[X_i]<\inf$. 

Entonces cuando $n \rightarrow \infty$:

$$ \frac{1}{n}\sum_{i=1}^{n}X_{i}\xrightarrow{c.s} \mu$$

Entonces, por la Ley fuerte de los grandes números, sabemos que un buen estimador del valor esperado de una variable aleatoria continua $X$ con función de distribucíon $f(x)$ es el valor promedio de una muestra finita (que, mientras más grande será mejor la aproximación) de variables aleatorias, independientes con distribución $f(x)$, es decir:

$$\frac{1}{n}\sum_{i=1}^{n}X_{i}\approx \mathbb{E}[X]$$

Como la esperanza de una variable aleatoria continua es una integral, entonces la media muestral también puede usarse para estimar el valor de una integral, en general:

$$E[X] = \int_{\alpha}^{\beta} x f(x) dx \approx \frac{1}{n}\sum_{i=1}^{n}X_{i}$$

## ¿En qué consiste el método Montecarlo?

¡La Ley Fuerte de los Grandes Números es la idea detrás del método Monte Carlo!

* Una sola simulación de Monte Carlo toma la variable que tiene incertidumbre y le asigna un valor aleatorio. 
* A continuación, se ejecuta el modelo y se proporciona un resultado. 
* Repetimos una y otra vez mientras se asigna a la variable aleatoria de interés muchos valores diferentes. 
* Una vez que se completan las simulaciones, los resultados se promedian juntos para proporcionar una estimación.

En nuestra área de interés, finanzas, es común es valuar instrumentos financieros cuyos rendimientos son aleatorios, como es el caso de los derivados. En la mayoría de los casos, la valuación se reduce al cálculo de una esperanza de una función continua.

El problema, es que frecuentemente no es posible aplicar un método de integración para calcular en forma exacta (analítica, para obtener una fórmula cerrada) la integral de la esperanza.

## Valuación de opciones por Método Montecarlo

En el caso de la valuación de opciones europeas, el método Monte-Carlo se usa para calcular la siguiente integral:

$$V_{t} = \mathbb{E}[e^{-r\Delta_{T}} h(S_{T})]$$

donde: 

* $t$: tiempo de valuación (En años)
* $T$: Tiempo de vencimiento de la opción (En años)
* $\Delta_{T}$ : $T - t$
* $S_{T}$: Precio del subyacente a tiempo T
* $h(S_{T})$: El payoff de la opción en función del precio del subyacente a la fecha de vencimiento

¡Muy bien! Pero... ¿Cómo modelamos $S_T$ al tiempo $t$?

Recordemos que, bajo la medida libre de riesgo $\mathbb{Q}$, tenemos que:

$$S_t = S_0e^{\left(r-\frac{\sigma^{2}}{2}\right)t+\sigma W_{t}}$$

Donde:

* $t$ es el tiempo de interés, al que queremos modelar el precio del activo subyacente
* $S_0$ es el precio del activo subyacente a tiempo 0 (El último precio observado en mercado)
* $r$ es la tasa libre de riesgo
* $\sigma$ es la volatilidad del activo subyacente
* $W_t$ es un movimiento browniano a tiempo $t$

¿Y cómo simulamos a $W_t$?

$W_t = Y\sqrt{t}$

Donde:

* $Y$ es una variable aleatoria normal estándar
* $t$ es el tiempo de interés, al que queremos modelar el precio del activo subyacente

Porque la volatilidad del browniano crece proporcional a la raíz cuadrada del tiempo

#### Entonces: ¿Cómo queda el modelo?

$$V_{t} = E\left[e^{-r\Delta_{T}} h\left(S_{t}e^{r- \frac{\sigma^{2}}{2}\Delta_{T} + \sigma Y \sqrt{\Delta_{T}}}\right)\right]$$

donde: 

* $T$: Tiempo de vencimiento de la opción
* $\Delta_{T}$ : T - t
* $S_{t}$: Precio del subyacente a tiempo t
* $h(S_{T})$: El payoff de la opción en función del precio del subyacente a la fecha de vencimiento
* $\sigma$: Volatilidad del subyacente
* $r$: Tasa libre de riesgo
* $Y$: Variable aleatoria con distribución normal con media 0 y varianza 1

#### Consideración: Coste computacional un poco alto

# Código para la simulación

## Librerías a usar

```python
import numpy as np #Libreria para vectores, operaciones matemáticas y distribuciones aleatorias
import pandas as pd #Libreria para manipulacion y visualización de dataframes
```

## Datos iniciales

```python
#Datos iniciales
T = 1 #Plazo de la opción (en años)
s0 = 50 #Precio del subyacente a día de valuación
r = 0.06 #Tasa libre de riesgo
sigma = 0.12 #Volatilidad del activo subyacente (Desviación estandar)
k = 52 #Strike (Precio de ejercicio)
sim_mont = 100000 #Número de simulaciones montecarlo
dt = 1/100 #Número de particiones en el plazo de la opcion
```

## Función para simular el precio del activo subyacente durante un intervalo $dt$

Aquí simularemos $S_t$ haciendo uso del modelo antes mencionado. 

$$S_t = S_{0}e^{r- \frac{\sigma^{2}}{2}\Delta_{T} + \sigma Y \sqrt{\Delta_{T}}}$$
* suma_1 = $r- \frac{\sigma^{2}}{2}\Delta_{T}$
* suma_2 = $\sigma Y \sqrt{\Delta_{T}}$
* $dt = \Delta_{t}$
* $Z = U$

```python
def precio_subyacente(s0, r, sigma, dt):
    Z = np.random.normal(0,1,1)[0] #SImulamos una distribución normal estándar
    suma_1 = (r - ((sigma ** 2)/ 2)) * dt 
    suma_2 = sigma * Z * np.sqrt(dt)
    precio_subyacente = s0 * np.exp(suma_1 + suma_2) #Calculamos el precio del subyacente
    return precio_subyacente
```

## Simulación de trayectorias del precio del activo subyacente

```python
lista_trayectorias = [] #Lista que va a guardar todas nuestras simulaciones

for it in range(0,sim_mont):
    lista_precios = [s0] #La simulacion a t=0 debe de ser igual al precio del subyacente al inicio de la opción
    st = s0 #Inicializamos variables
    #Cliclo FOR para calcular el precio del subyacente en cada particion
    for i in np.arange(0,T, dt):
        st = precio_subyacente(st, r, sigma, dt) #calculamos el precio cada dt periodos de tiempo
        lista_precios.append(st)
    lista_trayectorias.append(lista_precios) #agregamos precio del subyacente a la lista
```

```python
#Dataframe con la trayectorias del activo subyacente
#Cada columna representa una iteración
#Cada fila representa una partición
df_trayectorias = pd.DataFrame(lista_trayectorias, index = ["Iteracion " + str(i + 1) for i in range(sim_mont) ]).transpose()
df_trayectorias.tail()
```

## Gráfica con todas las trayectorias simuladas del activo subyacente

```python
df_trayectorias.plot(legend = False)
```

## Cáculo del payoff en cada trayectoria (con los precios al finales)

Aplicamos la función del payoff correspondiente a cada payoff

### Put

```python
df_trayectorias_put = df_trayectorias.iloc[-1,:].apply(lambda x: max(k - x, 0))
df_trayectorias_put
```

### Call

```python
df_trayectorias_call = df_trayectorias.iloc[-1,:].apply(lambda x: max(x - k, 0))
df_trayectorias_call
```

## Estimación del precio (Promedio de los payoffs de cada trayectoria)

### Put

```python
valor_MC_put = df_trayectorias_put.mean()
valor_MC_put
```

### Call

```python
valor_MC_call = df_trayectorias_call.mean()
valor_MC_call
```
